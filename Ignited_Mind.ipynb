**Outlier Detection**

```python
for feature in num_var:
    if feature != 'churn':
        sns.boxplot(x ='churn', y = feature, data = df)
        plt.title(feature)
        plt.show()
```

every features has a outliers so we need to remove the outliers.
outlies contains the some usefull information.
so we have to replace the outliers with some meaning full values. so we should replace the outliers with meadin values

**Removing the outliers**

```python
#functions for removing outliers
def remove_outliers(df,labels):
    for label in labels:
        q1 = df[label].quantile(0.25)
        q3 = df[label].quantile(0.75)
        iqr = q3 - q1
        upper_bound = q3 + 1.5 * iqr
        lower_bound = q1 - 1.5 * iqr
        df[label] = df[label].mask(df[label]< lower_bound, df[label].median(),axis=0)
        df[label] = df[label].mask(df[label]> upper_bound, df[label].median(),axis=0)

    return df
```

```python
df = remove_outliers(df, num_var)
```

```python
for feature in num_var:
    if feature != 'churn':
        sns.boxplot(x ='churn', y = feature, data = df)
        plt.title(feature)
        plt.show()
```
